```python
import networkx as nx
from operator import itemgetter
import matplotlib.pyplot as plt
import community.community_louvain as community_louvain #pip install python-louvain
```

```python
from MySNA import MySNA
```

```python
# sample graph
# you can also use the sample graphs provided in the directory
# G = nx.read_edgelist("connections.txt", create_using=nx.Graph(), nodetype = int)
# G = nx.read_edgelist("community.txt", create_using=nx.Graph(), nodetype = int)

G = nx.karate_club_graph()
```

```python
# draw the undirected graph
nx.draw(G, with_labels=True,  node_color='purple', edge_color='blue', font_color="whitesmoke")
ax = plt.gca()
ax.collections[0].set_edgecolor("black")
plt.show()
```

```python
# find node with the highest degree
print(MySNA.node_with_max_degree(G))
print("Node with maximum degree: ", MySNA.node_with_max_degree(G)[0])
print("Degree of the node with maximum degree: ", MySNA.node_with_max_degree(G)[1])
```

```python
# find the node with lowest degree
print(MySNA.node_with_min_degree(G))
print("Node with minimum degree: ", MySNA.node_with_min_degree(G)[0])
print("Degree of the node with minimum degree: ", MySNA.node_with_min_degree(G)[1])
```

```python
# Get the first 5 nodes with highest betweenness centrality, in decreasing order
MySNA.get_first_n_highest_betweenness_nodes(G, 5)
```

```python
# Get the first 3 nodes with highest closeness centrality, in decreasing order
MySNA.get_first_n_highest_closeness_nodes(G, 3)
```

```python
# Get the first 10 nodes with highest eigenvector centrality, in decreasing order
MySNA.get_first_n_highest_eigenvector_nodes(G, 10)
```

```python
# Find the communities using Louvain's 'best partition' method
MySNA.find_communities(G)
```

```python
# draw the communities
ax = MySNA.draw_communities(G)  # by default, label=True
plt.show()
```

```python
# find the influencers within each community
MySNA.find_community_influencers(G)
```

```python
# draw the graph, highlighting all the community influencers
ax = MySNA.draw_community_influencers(G)  #by default, label=True
plt.show()
```

```python
# find the community that contains the given node, e.g. '7'
MySNA.community_with_node(G, 7)
```

```python
# draw the community that contains node '7'
ax = MySNA.draw_community_with_node(G, 7)
plt.show()
```

```python
# get the summary of the problem
MySNA.problemsolver(G)
```
